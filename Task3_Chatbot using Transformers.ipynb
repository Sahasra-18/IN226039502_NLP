# **Chatbot:**
A chatbot is a computer program that can communicate with users in natural language, just like a human conversation. It takes user input, processes it, and generates a meaningful response.

In this project, a transformer-based model is used to build the chatbot. The model understands the input text and predicts the next words to generate a reply. The chatbot runs in a loop, allowing continuous interaction until the user exits.

It combines user input handling, response generation, and conversation flow to simulate a basic AI assistant.

# **3. Response Generation**

```python
from transformers import pipeline

chatbot = pipeline(
    "text-generation",
    model="gpt2",
    pad_token_id=50256   # ✅ removes warning
)
```

```python
while True:
    user_input = input("You: ")

    if user_input.lower() in ["exit", "quit"]:
        print("Chatbot: Goodbye!")
        break

    response = chatbot(user_input, max_new_tokens=50)

    reply = response[0]['generated_text']
    reply = reply.replace(user_input, "").strip()

    print("Chatbot:", reply)
```
